## Question One: First Duplicate in List
**Description**  
Given a list of integers, return the **first value that appears more than once**. If there are multiple duplicates, return the one that appears **first** in the list. If no duplicate exists, return `-1`.

**Examples**
```python
Input: nums = [3, 1, 4, 2, 5, 1, 6]
Output: 1
```
```python
Input: nums = [7, 8, 9, 10]
Output: -1
```
```python
Input: nums = [4, 5, 6, 4, 6]
Output: 4
```

**Question 1 Starter Code**

````python
from typing import List

def first_duplicate(nums: List[int]) -> int:
   # Create an empty set to store seen integers
	set_integers = set()

	for num in nums:
		if num in set_integers:
			# Found the first duplicate
			return num

		# Integer not found, add to the set
		set_integers.add(num)

	return -1

print(first_duplicate([4, 5, 6, 4, 6]))  # The output should be 4, since it's the first duplicate encountered
````


-   Explain why your solution works


The solution finds the first duplicate integer in a list by iterating through the list and using a set to keep track of integers that have already been seen. When it encounters an integer that is already in the set, it returns that integer as the first duplicate. If no duplicates are found, it returns -1.


-   Explain the problemâ€™s time and space complexity


````python
# The complexity for each statement is described in a comment beside, while the comment at the top of the function describes the overall time complexity

# Overall time complexity: O(n)
def first_duplicate(nums: List[int]) -> int:
   # Create an empty set to store seen integers
	set_integers = set() # O(1) - initializing an empty set

	for num in nums: # O(n) - iterating through the list of numbers
		if num in set_integers: # O(1) - checking membership in a set
			# Found the first duplicate
			return num # O(1) - returning a value

		# Integer not found, add to the set
		set_integers.add(num) # O(1) - adding an element to a set

	return -1 # O(1) - returning a value

# The output should be 4, since it's the first duplicate encountered
print(first_duplicate([4, 5, 6, 4, 6])) # O(1) - printing the result

# The output should be 2, since it's the first duplicate encountered
print(first_duplicate([1, 2, 3, 4, 5, 2, 3, 4, 5])) # O(1) - printing the result

# The output should be -6, since it's the first duplicate encountered
print(first_duplicate([-10, -9, -8, -7, -6, -6, -7, -8, -9, -10])) # O(1) - printing the result


# The complexity for each statement is described in a comment beside, while the comment at the top of the function describes the overall space complexity

# Overall space complexity: O(n)
def first_duplicate(nums: List[int]) -> int:
   # Create an empty set to store seen integers
	set_integers = set() # O(n) - in the worst case, we may store all unique integers in the set

	for num in nums: # O(n) - iterating through the list of numbers
		if num in set_integers: # O(1) - checking membership in a set
			# Found the first duplicate
			return num # O(1) - returning a value

		# Integer not found, add to the set
		set_integers.add(num) # O(1) - adding an element to a set

	return -1

# The output should be 4, since it's the first duplicate encountered
print(first_duplicate([4, 5, 6, 4, 6])) # O(1) - printing the result

# The output should be 2, since it's the first duplicate encountered
print(first_duplicate([1, 2, 3, 4, 5, 2, 3, 4, 5])) # O(1) - printing the result

# The output should be -6, since it's the first duplicate encountered
print(first_duplicate([-10, -9, -8, -7, -6, -6, -7, -8, -9, -10])) # O(1) - printing the result
````
